```python
import ollama

def ask_ollama(prompt):

    response = ollama.chat(
        model="qwen2.5:7b",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    return response["message"]["content"]
```

```python
import requests

def search_news(query, num_results=5):

    url = "https://google.serper.dev/search"

    headers = {
        "X-API-KEY": os.getenv("SERPER_API_KEY"),
        "Content-Type": "application/json"
    }

    payload = {
        "q": query,
        "num": num_results
    }

    response = requests.post(
        url,
        headers=headers,
        json=payload
    )

    return response.json()
```

```python
results = search_news(
    "latest AI stock market news"
)

print(results)
```

```python
def extract_news(results):

    news = []

    for item in results.get("organic", []):

        title = item.get("title", "")
        snippet = item.get("snippet", "")

        news.append(
            f"Title: {title}\nSummary: {snippet}"
        )

    return "\n\n".join(news)
```

```python
async def get_account_summary(name):

    async with stdio_client(server_params) as (read, write):

        async with ClientSession(read, write) as session:

            await session.initialize()

            balance = await session.call_tool(
                "get_balance",
                {"name": name}
            )

            holdings = await session.call_tool(
                "get_holdings",
                {"name": name}
            )

            return {
                "balance": balance.content[0].text,
                "holdings": holdings.content[0].text
            }
```

```python
server_params = StdioServerParameters(
    command="python3",
    args=["accounts_server.py"]
)

print(server_params)
```

```python
async def test_accounts_server():

    async with stdio_client(server_params) as (read, write):

        async with ClientSession(read, write) as session:

            await session.initialize()

            tools = await session.list_tools()

            return tools

tools = await test_accounts_server()

print(tools)
```

```python
def autonomous_trader(
    research_report,
    strategy,
    account_info
):

    prompt = f"""
You are an autonomous trading agent.

ACCOUNT INFORMATION:

{account_info}

INVESTMENT STRATEGY:

{strategy}

RESEARCH REPORT:

{research_report}

Based on the account balance, holdings,
investment strategy and research report:

1. Recommend stocks to buy
2. Recommend quantities
3. Explain rationale
4. Estimate risk
5. Give final action plan

Only recommend publicly tradable stocks.
"""

    return ask_ollama(prompt)
```

```python
import requests

def get_stock_price(symbol):

    api_key = os.getenv("POLYGON_API_KEY")

    url = (
        f"https://api.polygon.io/v2/aggs/ticker/"
        f"{symbol}/prev"
        f"?adjusted=true&apiKey={api_key}"
    )

    response = requests.get(url)

    return response.json()
```

```python
get_stock_price("AAPL")
```

```python
aapl_data = get_stock_price("AAPL")

print(aapl_data)
```

```python
def latest_price(symbol):

    data = get_stock_price(symbol)

    if (
        data.get("status") == "OK"
        and data.get("results")
    ):
        return data["results"][0]["c"]

    return None
```

```python
async def execute_buy(
    name,
    symbol,
    quantity,
    rationale
):

    async with stdio_client(server_params) as (read, write):

        async with ClientSession(read, write) as session:

            await session.initialize()

            result = await session.call_tool(
                "buy_shares",
                {
                    "name": name,
                    "symbol": symbol,
                    "quantity": quantity,
                    "rationale": rationale
                }
            )

            return result
```
